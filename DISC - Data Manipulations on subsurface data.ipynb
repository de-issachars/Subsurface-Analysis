```python
# import libraries
import numpy as np
import pandas as pd
import striplog
```

```python
df = pd.read_excel('DISC-Sample_mudlog_file.xlsx')
```

```python
df.head()
```

```python
df.info()
```

Fisrtly, we want to select all rows where shale is equal or greater than 70

```python
# we convert to float because we cannot use comparison operators between strings
# we also start from the second index because '%' cannot be converted into a float
df['SHALE'] = pd.to_numeric(df['SHALE'][1:])
df['SHALE']
```

```python
df1 = df[df['SHALE'] >= 70] 
df1
```

```python
df1.to_csv('Shale_great_70.csv', index = False)
```

Next we select all rows with shale equal to 100

```python
df2 = df[df['SHALE'] == 100]
df2
```

```python
df2.to_csv('Shale_equals_100.csv', index = False)
```

Now we want to select all rows where the description is Cement

```python
df['DESCRIPTION'] = df['DESCRIPTION'].astype(str)
```

```python
def Cement_string(description):
    if 'Cement' in description:
        return True
    else:
        return False
```

```python
df3 = df[df['DESCRIPTION'].apply(lambda x:Cement_string(x))]
df3
```

```python
df3.to_csv('Cement_decriptions.csv', index = False)
```

Now we want to expand the description of all rows with Shale greater than 70

```python
df1['DESCRIPTION'] =df1['DESCRIPTION'].astype(str)
```

```python
from striplog import Lexicon
```

```python
Lexicon.default()
```

```python
def expand_abbreviations(description):
    return Lexicon.default().expand_abbreviations(description)
```

```python
df1['DESCRIPTION_FULL'] = df1['DESCRIPTION'].apply(lambda x:expand_abbreviations(x))
```

```python
df1
```

```python
df1.to_csv('Shale_full_description.csv', index = False)
```
