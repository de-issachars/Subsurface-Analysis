```python
logs = las.read('DISC-Sample_Las_file.LAS')
```

```python
data = logs.df()
data
```

```python
# Cross plot comparing GR to Resistivity
plt.figure(figsize=(5,4), dpi=100) # figure size and dpi you can set here
plt.scatter(data['GR_ARC'], data['A40H_UNC'], color='blue', marker='.', alpha=0.3)
plt.yscale('log') #log scale for Y axis
plt.grid(True)
plt.xlabel('GR', size=16)
plt.ylabel('Deep Resistivity', size=16)
```

```python
# lets make a histogram of the GR Curve, with 20 bins, in green
plt.hist(data['GR_ARC'], bins = 20, color='g', alpha = 0.5)
plt.xlabel('GR', size=16)
plt.ylabel('Count', size=16)
```

```python
# lets make a histogram of the Resistivity Curve, with 20 bins, in green
plt.hist(data['A40H_UNC'], bins = 20, color='g', alpha = 0.5)
plt.xlabel('Resistivity', size=16)
plt.ylabel('Count', size=16)
```

```python
# line plot of the Gamma curve
plt.figure(figsize=(4,8), dpi=100)
plt.plot(data.GR_ARC.values, data.index, color='g')
plt.ylabel('Depth in Feet', size=16)
plt.ylim(15000,8000)
plt.grid(True)
```

```python
# line plot of the Gamma curve
plt.figure(figsize=(6,7), dpi=100)

plt.suptitle('Well Sample') # overall title

plt.subplot(121) # if we are going to make two plots, matplotlib calls it a subplot
plt.plot(data.GR_ARC.values, data.index, color='g') # the actual plot!
plt.ylabel('Depth in Feet', size=16)
plt.xlabel('API', size=16)
plt.ylim(15428,7900) # the limit is reversed to go deep down
plt.grid(True) # Turning the grid on

plt.subplot(122)
plt.plot(data.A40H_UNC.values, data.index, color='grey')
plt.xscale('log')
plt.grid(True)
plt.xlim(0.4,4000)
plt.ylim(15428,7900)
plt.xlabel('OhmM', size=16)
plt.tick_params(labelleft=False)  

plt.savefig('awesome-plot.pdf', dpi=200)
```

We can set cut-offs that we would use to create a payflag between the two plots. The payflag will demonstrate instances where certain condition are met.

```python
lowGR = 80
highRES= 5
```

```python
payMapper = (data.GR_ARC.values <= lowGR) & (data.A40H_UNC.values >= highRES)
np.unique(payMapper)
```

```python
# line plot of the Gamma curve
plt.figure(figsize=(5,8), dpi=100)

plt.subplot(131) # the subplot changed to allow for 3 columns
plt.plot(data.GR_ARC.values, data.index, color='g')
plt.ylabel('Depth in Feet', size=16)
plt.xlabel('API', size=16)
plt.ylim(14500,14400)
plt.grid(True)

plt.subplot(132) # this is the new pay flag plot
plt.fill_between(payMapper, data.index, color='red') # used fill between, not plot
plt.ylim(14500,14400)
plt.tick_params(labelleft=False)  
plt.xlabel('Pay Flag', size=16)
plt.grid(True)

plt.subplot(133)
plt.plot(data.A40H_UNC.values, data.index, color='black')
plt.xscale('log')
plt.grid(True)
plt.xlim(0.4,4000)
plt.ylim(14500,14400)
plt.xlabel('OhmM', size=16)
plt.tick_params(labelleft=False)  

```
